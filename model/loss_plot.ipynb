```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open("./wyniki/preds.npy", "rb") as f:
    preds = np.load(f)
with open("./wyniki/trains.npy", "rb") as f:
    trains = np.load(f)
with open("./wyniki/trues.npy", "rb") as f:
    trues = np.load(f)
with open("./wyniki/vals.npy", "rb") as f:
    vals = np.load(f)
```

```python
def plot_results(trains, vals, preds, trues, n_samples=200):
    """Plot training curves and forecast vs actual."""
    fig, axes = plt.subplots(1, 2, figsize=(14, 4))

    # Loss curves
    ax = axes[0]
    ax.plot(trains, label="Train loss")
    ax.plot(vals,   label="Val loss")
    ax.set_title("Training & Validation Loss")
    ax.set_xlabel("Epoch"); ax.set_ylabel("MSE")
    ax.legend(); ax.grid(True, alpha=0.3)

    # Forecast vs actual (flatten first horizon step across samples)
    ax = axes[1]
    y_hat = preds[:n_samples, 0]
    y_true = trues[:n_samples, 0]
    ax.plot(y_true, label="Actual",    alpha=0.8)
    ax.plot(y_hat,  label="Predicted", alpha=0.8)
    ax.set_title("Forecast vs Actual (first horizon step)")
    ax.set_xlabel("Sample"); ax.set_ylabel("Wind Production")
    ax.legend(); ax.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig("tsmixer_results2.png", dpi=150)
    plt.show()
    print("Plot saved to tsmixer_results.png")
```

```python
plot_results(trains,vals, preds, trues)
```
